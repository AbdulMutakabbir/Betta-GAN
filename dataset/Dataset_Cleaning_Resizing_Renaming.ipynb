# Imports

```python
import os
import cv2
```

# Constants

```python
ORIGINAL_DATASET_V1_PATH = "Original_Betta_V1"
CROPED_DATASET_V1_PATH = "Cropped_Betta_V1"
CROPED_60X60_DATASET_V1_PATH = "Cropped_60x60_Betta_V1"
CROPED_600X600_DATASET_V1_PATH = "Cropped_600x600_Betta_V1"
CROPED_FLIPPED_60X60_DATASET_V1_PATH = "Cropped_Flipped_60x60_Betta_V1"
CROPED_FLIPPED_600X600_DATASET_V1_PATH = "Cropped_Flipped_600x600_Betta_V1"
ORIGINAL_DATASET_V2_PATH = "Original_Betta_V2"
CROPED_DATASET_V2_PATH = "Cropped_Betta_V2"
CROPED_60X60_DATASET_V2_PATH = "Cropped_60x60_Betta_V2"
CROPED_600X600_DATASET_V2_PATH = "Cropped_600x600_Betta_V2"
CROPED_FLIPPED_60X60_DATASET_V2_PATH = "Cropped_Flipped_60x60_Betta_V2"
CROPED_FLIPPED_600X600_DATASET_V2_PATH = "Cropped_Flipped_600x600_Betta_V2"
```

# Global Functions

```python
def get_image_list(directory):
    return os.listdir(directory)
```

# Rename
##### All the images in the directory will be name according to a standard convention

```python
def rename_images(directory):
    image_list = get_image_list(directory)
    print("Starting renaming process for directory: " + directory + "...")
    for index, image in enumerate(image_list):
        image_path = directory + os.sep + image
        image_root, image_extension = os.path.splitext(image_path)
        new_image_name = directory.upper() + "_" + str(index)
        new_image_path = directory + os.sep + new_image_name + image_extension
        os.rename(image_path, new_image_path)
        print("\tFile: "+ image + " Renamed: " + new_image_name)
    print("Renaming process completed.")
```

```python
# remane croped images version 1 
rename_images(CROPED_DATASET_V1_PATH)
```

```python
# rename original images version 1
rename_images(ORIGINAL_DATASET_V1_PATH)
```

```python
# remane croped images version 2 
rename_images(CROPED_DATASET_V2_PATH)
```

```python
# rename original images version 2
rename_images(ORIGINAL_DATASET_V2_PATH)
```

# Resize Images

```python
IMAGE_SIZE_60X60 = (60, 60)
IMAGE_SIZE_600X600 = (600, 600)
```

```python
def resize_images(directory, new_directory, size):
    image_list = get_image_list(directory)
    print("Starting resizing process for directory: " + directory + "...")
    for index, image in enumerate(image_list):
        image_path = directory + os.sep + image
        original_image = cv2.imread(image_path)
        resized_image = cv2.resize(original_image, size)
        resized_image_name = new_directory + "_" + image.split("_")[-1]
        resized_image_path = new_directory + os.sep + resized_image_name
        cv2.imwrite(resized_image_path, resized_image)
        print("\tFile: "+ image + " Resized: " + resized_image_name)
    print("Renaming process completed. Resized files in \"" + new_directory + "\".")
```

```python
# resize croped images version 1 to 600x600
resize_images(CROPED_DATASET_V1_PATH, CROPED_600X600_DATASET_V1_PATH, IMAGE_SIZE_600X600)
```

```python
# resize croped images version 1 to 60x60
resize_images(CROPED_DATASET_V1_PATH, CROPED_60X60_DATASET_V1_PATH, IMAGE_SIZE_60X60)
```

```python
# resize croped images version 2 to 600x600
resize_images(CROPED_DATASET_V2_PATH, CROPED_600X600_DATASET_V2_PATH, IMAGE_SIZE_600X600)
```

```python
# resize croped images version 2 to 60x60
resize_images(CROPED_DATASET_V2_PATH, CROPED_60X60_DATASET_V2_PATH, IMAGE_SIZE_60X60)
```

# Flip Images

```python
X_AXIES = 0
Y_AXIES = 1
```

```python
def flip_images(directory, new_directory, axies):
    image_list = get_image_list(directory)
    print("Starting image flip along " + str(axies) + " axies for directory: " + directory + "...")
    for index, image in enumerate(image_list):
        image_path = directory + os.sep + image
        original_image = cv2.imread(image_path)
        flipped_image = cv2.flip(original_image, axies)
        flipped_image_name = directory + "_FLIP_" + str(axies) + "_" + image.split("_")[-1]
        flipped_image_path = new_directory + os.sep + flipped_image_name
        cv2.imwrite(flipped_image_path, flipped_image)
        print("\tFile: "+ image + " Resized: " + flipped_image_name)
    print("Process completed. Fliped files in \"" + new_directory + "\".")
```

```python
# Flip 600x600 version 1 images horizontally
flip_images(CROPED_600X600_DATASET_V1_PATH, CROPED_FLIPPED_600X600_DATASET_V1_PATH, Y_AXIES)
```

```python
# Flip 60x60 version 1 images horizontally
flip_images(CROPED_60X60_DATASET_V1_PATH, CROPED_FLIPPED_60X60_DATASET_V1_PATH, Y_AXIES)
```

```python
# Flip 600x600 version 2 images horizontally
flip_images(CROPED_600X600_DATASET_V2_PATH, CROPED_FLIPPED_600X600_DATASET_V2_PATH, Y_AXIES)
```
